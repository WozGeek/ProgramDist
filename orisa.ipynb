```python
from transformers import pipeline
# Initialiser le pipeline avec Phi-3-mini
nlp = pipeline("text-generation", model="microsoft/Phi-3-mini-4k-instruct")
# Question test
question = "What are the main economic factors driving inflation in Africa in 2025?"
# Générer la réponse
response = nlp(question, max_length=200, truncation=True, num_return_sequences=1)[0]["generated_text"]
print(f"Question : {question}")
print(f"Réponse : {response}")
```

```python
# Liste de questions
questions = [
    "What are the main economic factors driving inflation in Africa in 2025?",
    "Why are prices rising in 2025?",
    "What impacts economic growth in Africa?",
    "How can students analyze economic trends?",
    "What are the geopolitical risks in Africa in 2025?"
]
# Tester chaque question
for question in questions:
    response = nlp(question, max_length=200, truncation=True, num_return_sequences=1)[0]["generated_text"]
    print(f"\nQuestion : {question}")
    print(f"Réponse : {response}\n{'-'*50}")
```

```python
from transformers import pipeline
import ipywidgets as widgets
from IPython.display import display
# Charger Phi-3-mini
nlp = pipeline("text-generation", model="microsoft/Phi-3-mini-4k-instruct", torch_dtype="float16")
print("Modèle Phi-3 chargé !")
# Créer une interface
question_input = widgets.Text(
    value="What are the main economic factors driving inflation in Africa in 2025?",
    description="Question :",
    layout={'width': '80%'}
)
output = widgets.Output()
def on_button_clicked(b):
    with output:
        output.clear_output()
        question = question_input.value
        if question:
            response = nlp(question, max_length=200, truncation=True, num_return_sequences=1)[0]["generated_text"]
            print(f"Question : {question}")
            print(f"Réponse : {response}")
        else:
            print("Entrez une question !")
button = widgets.Button(description="Analyze with ORISA")
button.on_click(on_button_clicked)
display(question_input, button, output)
```
